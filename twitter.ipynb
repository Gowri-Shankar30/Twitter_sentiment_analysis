```python
import pandas as pd
import re
from nltk.stem import PorterStemmer

# 1. Load the dataset
dataset = pd.read_csv("training.1600000.processed.noemoticon.csv", encoding='ISO-8859-1', header=None)
dataset.columns = ['target', 'ids', 'date', 'flag', 'user', 'text']

# 2. Handle missing values (if any)
dataset['text'] = dataset['text'].fillna('')

# 3. Initialize stemmer
stemmer = PorterStemmer()

# 4. Define stemming function
def stemming(content):
    content = re.sub('[^a-zA-Z]', ' ', str(content))  # Remove non-alphabetic chars
    content = content.lower().split()  # Lowercase and split
    content = [stemmer.stem(word) for word in content]  # Stem each word
    return ' '.join(content)

# 5. Apply stemming
dataset['text'] = dataset['text'].apply(stemming)

# 6. Check results
print(dataset['text'].head())
```

```python
dataset.head()
```

```python
x = dataset['text']
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split  # <-- Add this import

# Assuming x and y are already defined
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer  # <-- Add this import

# Convert textual data to numerical data
vectorizer = TfidfVectorizer()
x_train = vectorizer.fit_transform(x_train)  # Learn vocabulary and idf, return term-document matrix
x_test = vectorizer.transform(x_test)  # Transform documents to document-term matrix
```

```python
print(x_train)
```

```python
from sklearn.linear_model import LogisticRegression  # <-- Add this import

# Training the model
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Testing the model
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
from nltk.corpus import stopwords
def predict_sentiment(text):
    text = re.sub('[^a-zA-Z]',' ',text) # removing not a-z and A-Z
    text = text.lower()
    text = text.split() 
    text = [stemmer.stem(word) for word in text if not word in stopwords.words('english')]
    text = ' '.join(text)
    text = [text]
    text = vectorizer.transform(text)   
    sentiment = model.predict(text)
    if sentiment == 0:
        return "Negative"
    else:
        return "Positive"
```

```python
# Testing the model
print(predict_sentiment("I hate you"))
print(predict_sentiment("I love you"))
```
